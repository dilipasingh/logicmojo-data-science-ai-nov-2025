Write a function `second_largest(numbers)` that takes a list of numbers and returns the second largest number in the list. Assume the list has at least two unique numbers.

Example:
```python
numbers = [10, 5, 20, 8, 15]
```

Write a Python program to print all the prime numbers between 1 and N (inclusive). Where N is a positive integer and an input from user.

Example:

```python
N = 20
# Expected output: 2,3,5,7,11,13,17,19

```

````python
# Write your code here
# I copied this code unable too solve it.


def print_primes(n):
    """
    Prints all prime numbers between 1 and n (inclusive).
    Uses the Sieve of Eratosthenes algorithm for efficiency.
    
    Args:
        n: Upper limit (inclusive)
    """
    if n < 2:
        print("No prime numbers in this range.")
        return
    
    # Create a boolean array "is_prime" and initialize all entries as True
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False  # 0 and 1 are not prime
    
    # Sieve of Eratosthenes
    p = 2
    while p * p <= n:
        if is_prime[p]:
            # Mark all multiples of p as not prime
            for i in range(p * p, n + 1, p):
                is_prime[i] = False
        p += 1
    
    # Print all prime numbers
    primes = [num for num in range(2, n + 1) if is_prime[num]]
    print(f"Prime numbers between 1 and {n}:")
    print(primes)
    print(f"\nTotal count: {len(primes)}")

# Example usage
num = int(input(f'Enter a positive number range: '))
print(print_primes(num))
````

#### Question-4

Write a function `merge_sorted_lists(list1, list2)` that takes two sorted lists of integers, `list1` and `list2` and merges them into a single sorted list efficiently. One way is to simple concatenate the lists and sort. But think of a better approach.

Example:
```python
list1 = [1, 4, 5]
list2 = [1, 3, 4]
merged = merge_sorted_lists(list1, list2)
print(merged)
# Expected output: [1, 1, 3, 4, 4, 5]
```

````python
# Write your code here
def merge_sorted_lists(list1, list2):
    list1.extend(list2)
    list1.sort()
    return list1
````

#### Question-6

Write a function `is_anagram(str1, str2)` that checks if two given strings are anagrams of each other (contain the same characters with the same frequencies, ignoring case).

Example:
```python
print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
# Expected output:
# True
# False
```

````python
# Write your code here
def is_anagram(str1,str2):
    count=0
    if len(str1) == len(str2):
        for s in str1:
            if s.upper() in str2.upper():
                count+=1
            else:
                continue
    if count==len(str1) and count==len(str2):
        return True
    else:
        return False
````

````python
print(is_anagram("Listen", "silEnt"))
print(is_anagram("hello", "world"))
````

Write a Python program that prints the first `n` rows of Pascal's triangle. `n` is a positive integer and an user input. To know about Pascal's triangle see the [wikipedia article](https://en.wikipedia.org/wiki/Pascal%27s_triangle): 

Example:

```python
n = 5
# Expected output:
#     1
#    1 1
#   1 2 1
#  1 3 3 1
# 1 4 6 4 1
```

````python
# Write your code here
# I copied this code unable too solve it.

def print_pascals_triangle(n):
    """
    Prints the first n rows of Pascal's triangle.
    
    Args:
        n: Number of rows to print
    """
    for i in range(n):
        # Print leading spaces for formatting
        print(' ' * (n - i - 1), end='')
        
        # Calculate and print values for this row
        num = 1
        for j in range(i + 1):
            print(num, end=' ')
            # Calculate next value using: C(n,k+1) = C(n,k) * (n-k) / (k+1)
            num = num * (i - j) // (j + 1)
        
        print()  # New line after each row

# Example usage
print_pascals_triangle(5)
````
